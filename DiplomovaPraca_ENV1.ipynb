```python
import re
import requests
import unicodedata
import bs4

def getParagraphs(page):

    r = requests.get('https://en.wikipedia.org/wiki/{0}'.format(page))
    soup = bs4.BeautifulSoup(r.content)
    paragraphs = soup.find_all('p')

    for p in paragraphs:
        cleanedText = re.sub('(\[[0-9]+\])', '', unicodedata.normalize('NFKD', p.text)).strip()
        if cleanedText:
            yield cleanedText

text = list(getParagraphs('Romeo_and_Juliet'))[9]
```
